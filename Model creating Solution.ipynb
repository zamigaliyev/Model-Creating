#### Find Q1 and Q3

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
IQR
```

#### Remove outliers of price column and call it as 'df_filtered'

```python
df_filtered = df[(df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))]
df_filtered
```

```python
df_filtered_out = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df_filtered_out 
```

# Multiple Regression

Libraries for Multiple Regression

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

#### Define X as carat, depth, table and y as price column and find accuracy; errors

```python
X = df_filtered_out[['carat','depth','table']]
y = df_filtered_out.iloc[:,7]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train)
```

```python
coeff_df = pd.DataFrame(regressor.coef_, X.columns, columns=['Coefficient'])  
coeff_df
```

```python
y_pred = regressor.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df1 = df.head(25)
```

```python
df1.plot(kind='bar',figsize=(10,8))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
accuracy = regressor.score(X_test,y_test)
accuracy
```

```python
plt.scatter(y_test, y_pred, color='r')
plt.show()
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

# Decision Tree

#### Define X as carat, depth, table,price and y as cut column and find accuracy

```python
df_filtered_out.head()
```

```python
X = df_filtered_out[['depth','carat','table','price']]
y = df_filtered_out.iloc[:,1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

# Random Forest

#### Define X as price, depth, table and y as carat column and find accuracy; Mean Absolute Errors

```python
X = df_filtered_out[['price','depth','table']]
y = df_filtered_out.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state = None)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(n_estimators = 10, random_state = None)

rf.fit(X_train, y_train);
```

```python
y_predict = rf.predict(X_test)
```

```python
errors = abs(y_predict - y_test)

print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
```

```python
mape = 100 * (errors / y_test)

accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```
